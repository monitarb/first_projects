```python
import pandas as pd
import folium as fl
from folium import plugins
import pandas as pd
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
#Read top 15 stations file. With Coordiates, Title and Color
station_coords=pd.read_csv('https://raw.githubusercontent.com/smeetvikani/dsp/master/resources/station_coords_v2.csv', index_col=0).reset_index()
```

```python
station_coords.head()
```

```python
#Below ENTRIES_Multiple refers to the Radius of the circle on the map. 
ENTRIES=[math.log(i+1)-15 for i in station_coords.Entry_Count.tolist()]
ENTRIES_Multiple=[x*30 for x in ENTRIES]
```

```python
#Convert Station Coordiates into a list for folium. 
station_coords_list=[eval(i) for i in station_coords.Coordinates.tolist()]
```

```python
#Convert Station Colors(that would appear on the map) into a list for folium. 
station_colors=station_coords.color.tolist()
```

```python
#Set default map for that would appear when map object is called. 
default_map_loc=[40.730610,-73.935242]
```

```python
#Initiate the Map Object with specific attributes from the CSV. 
map2 = fl.Map(location=default_map_loc, zoom_start=12, tiles='cartodbpositron')

for i in range(0,len(station_coords_list)):
    fl.Marker(station_coords_list[i],popup=str(station_coords.Station_Name[i]),icon=fl.Icon(color=station_colors[i])).add_to(map2)
    fl.CircleMarker(station_coords_list[i],
                    radius=ENTRIES_Multiple[i],
                    color=station_colors[i],
                    fill_color=station_colors[i]
                    ).add_to(map2)
    
map2
```
